```python
import matplotlib.pyplot as plt
from PIL import Image
from pathlib import Path
```

```python
def display_images(directory="IMGs"):
    """
    Display all images in the specified directory.
    Supported formats: jpg, jpeg, png, bmp, gif, webp
    """
    image_extensions = {".jpg", ".jpeg", ".png", ".bmp", ".gif", ".webp"}

    image_paths = sorted(
        p for p in Path(directory).iterdir()
        if p.is_file() and p.suffix.lower() in image_extensions
    )

    if not image_paths:
        print(f"No images found in '{directory}'.")
        return

    for path in image_paths:
        img = Image.open(path)

        plt.figure(figsize=(6, 6))
        plt.imshow(img)
        plt.title(path.name)
        plt.axis("off")
        plt.show()
```

```python
import numpy as np
from typing import Union
def binarize(mask: Union[list, np.ndarray], threshold: float = 0.5) -> np.ndarray:
    """Convert probabilities or grayscale masks to binary masks."""
    # 1. Convert input to a numpy array (if it isn't already)
    arr = np.asarray(mask)

    # 2. Safety check: Make sure the threshold is actually a valid percentage
    if not (0.0 <= threshold <= 1.0):
        raise ValueError("Threshold must be between 0.0 and 1.0.")
    
    # 3. The core logic
    return (arr >= threshold).astype(np.uint8)
```

```python
binarize([1,2,3,4,5,9])
```
