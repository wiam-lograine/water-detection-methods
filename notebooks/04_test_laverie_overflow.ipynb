# 04 - Test sur les images locales et estimation du debordement

Objectif : appliquer le modele choisi aux images de `IMGs`, puis calculer un score de confiance lie au debordement.

```python
import sys
from pathlib import Path

ROOT = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()
sys.path.append(str(ROOT / "src"))

from water_detection_methods.data import list_local_images, load_image
from water_detection_methods.overflow import overflow_confidence, overflow_label, rectangular_zone
from water_detection_methods.visualization import overlay_mask, show_image

import matplotlib.pyplot as plt
import numpy as np
```

```python
local_images = list_local_images(ROOT / "IMGs")
print(f"Images locales : {len(local_images)}")
local_images
```

```python
image = load_image(local_images[0], size=(384, 256))

# A remplacer par la prediction du modele final.
predicted_water_mask = np.zeros(image.shape[:2], dtype=np.uint8)

# Exemple : zone critique en bas de l'image. A adapter pour chaque camera.
critical_zone = rectangular_zone(image.shape[:2], x_min=0.0, y_min=0.75, x_max=1.0, y_max=1.0)
result = overflow_confidence(predicted_water_mask, critical_zone, alert_ratio=0.15)
label = overflow_label(result["overflow_confidence"])

print(result)
print("Decision:", label)

fig, axes = plt.subplots(1, 2, figsize=(11, 4))
show_image(image, "Image locale", axes[0])
show_image(overlay_mask(image, critical_zone, color=(1.0, 0.0, 0.0), alpha=0.25), "Zone critique", axes[1])
plt.tight_layout()
```
