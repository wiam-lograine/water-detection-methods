Objectif : entrainer un modele de segmentation semantique pour predire un masque eau / non-eau.

```python
import sys
from pathlib import Path

ROOT = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()
sys.path.append(str(ROOT / "src"))

from water_detection_methods.data import find_water_v2_pairs, train_val_split, load_pair
from water_detection_methods.visualization import show_image_mask_overlay
```

```python
pairs = find_water_v2_pairs(ROOT / "water_v2")
train_pairs, val_pairs = train_val_split(pairs, val_ratio=0.2, seed=42)

print(f"Train: {len(train_pairs)}")
print(f"Validation: {len(val_pairs)}")

image, mask = load_pair(*train_pairs[0], size=(256, 256))
show_image_mask_overlay(image, mask, title="Exemple d'entrainement")
```
